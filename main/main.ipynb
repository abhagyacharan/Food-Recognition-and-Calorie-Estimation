**Combining datasets**

```python
import os
import shutil
from random import sample

def combine_datasets(dataset1_path, dataset2_path, output_dataset_path, max_images_per_class):
    # Ensure output path exists
    os.makedirs(output_dataset_path, exist_ok=True)

    # Function to get class names from a dataset path
    def get_class_names(dataset_path):
        return [name for name in os.listdir(dataset_path) if os.path.isdir(os.path.join(dataset_path, name))]

    # Get class names from both datasets and ensure uniqueness
    dataset1_classes = get_class_names(dataset1_path)
    dataset2_classes = get_class_names(dataset2_path)
    combined_classes = list(set(dataset1_classes + dataset2_classes))  # Combine and ensure uniqueness

    for class_name in combined_classes:
        output_class_path = os.path.join(output_dataset_path, class_name)
        os.makedirs(output_class_path, exist_ok=True)  # Ensure output class directory exists

        # Handle Classes from Both Datasets
        for dataset_path, dataset_name in [(dataset1_path, 'Dataset 1'), (dataset2_path, 'Dataset 2')]:
            class_path = os.path.join(dataset_path, class_name)
            if os.path.exists(class_path):
                images = [f for f in os.listdir(class_path) if f.endswith('.jpg') or f.endswith('.png')]
                images_to_move = images[:max_images_per_class]  # Limit to max_images_per_class
                for image in images_to_move:
                    shutil.move(os.path.join(class_path, image), output_class_path)

                # If not enough images in the first dataset, fill up to max_images_per_class from the second dataset
                remaining_slots = max_images_per_class - len(os.listdir(output_class_path))
                if remaining_slots > 0:
                    additional_images = sample([img for img in images if img not in images_to_move], min(remaining_slots, len(images)))
                    for image in additional_images:
                        shutil.move(os.path.join(class_path, image), output_class_path)

# Example usage
if __name__ == "__main__":
    combine_datasets(
        dataset1_path="D:/Github/Datasets/Indian Food Images/Indian Food Images",
        dataset2_path="D:/Github/Datasets/indian20/dataset/Dataset/train",
        output_dataset_path="D:/Github/Datasets/combined_indian_dataset",
        max_images_per_class=50
    )
```

**Preprocess and Augmentation**

```python
def preprocess(img_paths):
    images = []
    for img_path in img_paths:
        img_data = cv2.imread(img_path)
        img_rgb = cv2.cvtColor(img_data, cv2.COLOR_BGR2RGB)
        image_resized = cv2.resize(img_rgb, (224, 224))
        image_normalized = image_resized.astype('float32') / 255
        images.append(image_normalized)

    return np.array(images)
```

```python
#augmention
datagen_train = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2, #horizontal shift
    height_shift_range=0.2, #vertical shift
    shear_range=0.2,
    zoom_range=0.2, #random zoom
    horizontal_flip=True, #random horizontal flips
    fill_mode='nearest' 
)
datagen_val = ImageDataGenerator(
    rescale=1./255 
)
```

```python
train_dataset = datagen_train.flow_from_dataframe(
    dataframe=pd.DataFrame({'img_path':x_train,'label':y_train}),
    x_col='img_path',
    y_col='label',
    target_size=(224, 224),  
    batch_size=32,
    class_mode='categorical', 
    shuffle=True,
    seed=42 
)

validation_dataset = datagen_val.flow_from_dataframe(
    dataframe=pd.DataFrame({'img_path':x_val,'label':y_val}),
    x_col='img_path',
    y_col='label',
    target_size=(224, 224),  
    batch_size=32,
    class_mode='categorical', 
    shuffle=True,
    seed=42 
)
```

**Model Training**

```python
def build_model(base):
    base_model = base
    base_model.trainable = False  # Freeze the base model

    model = Sequential([
        base_model,
        Flatten(),
        
        Dense(512, activation='relu'),
        Dropout(0.25),
        BatchNormalization(),

        Dense(256, activation='relu'),
        Dropout(0.25),
        BatchNormalization(),
        
        Dense(128, activation='relu'),
        Dropout(0.25),
        BatchNormalization(),
        
        Dense(96, activation='softmax'),  # For binary classification
    ])

    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from keras.models import load_model
from sklearn.metrics import confusion_matrix, classification_report
from keras.preprocessing import image
from keras.utils import load_img, img_to_array
from keras.applications.mobilenet_v2 import preprocess_input

def analyze_model_performance(model, validation_dataset, class_indices):
    """
    Analyze and visualize per-class performance of the food recognition model.
    
    Parameters:
    model: trained MobileNetV2 model
    validation_dataset: ImageDataGenerator validation dataset
    class_indices: dictionary mapping class names to indices
    
    Returns:
    dict: Dictionary containing performance metrics per class
    """
    # Reset the validation dataset
    validation_dataset.reset()
    
    # Get predictions and true labels
    y_pred = []
    y_true = []
    
    # Create reverse mapping from index to class name
    index_to_class = {v: k for k, v in class_indices.items()}
    
    # Get predictions for all validation images
    steps = validation_dataset.samples // validation_dataset.batch_size + 1
    for i in range(steps):
        x_batch, y_batch = validation_dataset.next()
        
        # Get predictions for batch
        predictions = model.predict(x_batch)
        pred_indices = np.argmax(predictions, axis=1)
        true_indices = np.argmax(y_batch, axis=1)
        
        y_pred.extend(pred_indices)
        y_true.extend(true_indices)
        
        if len(y_true) >= validation_dataset.samples:
            break
    
    # Trim to actual number of samples
    y_pred = y_pred[:validation_dataset.samples]
    y_true = y_true[:validation_dataset.samples]
    
    # Convert indices to class names
    y_pred_classes = [index_to_class[idx] for idx in y_pred]
    y_true_classes = [index_to_class[idx] for idx in y_true]
    
    # Calculate metrics
    report = classification_report(y_true_classes, y_pred_classes, output_dict=True)
    df_report = pd.DataFrame(report).transpose()
    
    # Sort classes by performance
    class_performance = df_report.iloc[:-3]  # Remove avg/total rows
    class_performance = class_performance.sort_values('f1-score')
    
    # Plotting F1 Scores
    plt.figure(figsize=(15, 10))
    sns.barplot(x=class_performance['f1-score'], 
                y=class_performance.index)
    plt.title('F1 Score by Food Class')
    plt.xlabel('F1 Score')
    plt.ylabel('Food Class')
    plt.tight_layout()
    plt.show()
    
    # Confusion Matrix for worst performing classes
    worst_classes = class_performance.nsmallest(10, 'f1-score').index.tolist()
    worst_class_indices = [class_indices[cls] for cls in worst_classes]
    
    # Get corresponding true and predicted values for worst classes
    mask_true = np.isin(y_true, worst_class_indices)
    mask_pred = np.isin(y_pred, worst_class_indices)
    mask = mask_true | mask_pred
    
    y_true_worst = [index_to_class[idx] for idx in np.array(y_true)[mask]]
    y_pred_worst = [index_to_class[idx] for idx in np.array(y_pred)[mask]]
    
    cm_worst = confusion_matrix(y_true_worst, y_pred_worst, labels=worst_classes)
    
    plt.figure(figsize=(12, 12))
    sns.heatmap(cm_worst, annot=True, fmt='d', 
                xticklabels=worst_classes,
                yticklabels=worst_classes)
    plt.title('Confusion Matrix for 30 Worst Performing Classes')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()
    
    # Print analysis
    print("\nWorst Performing Food Classes:")
    worst_classes_stats = class_performance.nsmallest(30, 'f1-score')
    print(worst_classes_stats[['precision', 'recall', 'f1-score', 'support']])
    
    print("\nBest Performing Food Classes:")
    best_classes_stats = class_performance.nlargest(20, 'f1-score')
    print(best_classes_stats[['precision', 'recall', 'f1-score', 'support']])
    
    return {
        'class_performance': class_performance,
        'worst_classes': worst_classes,
        'best_classes': best_classes_stats.index.tolist()
    }
```
